### Helps detect unusually high purchase values.

```python

prices = data['Purchase Price'].values
z_scores = (prices - np.mean(prices)) / np.std(prices)

outliers = prices[np.abs(z_scores) > 3]
print("Outliers detected:", len(outliers))
```
